```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os, netCDF4, xarray

os.chdir('/Users/hausfath/Desktop/Climate Science/Stripe/Net zero GHG CDR/')
```

```python
CH4GWP100 = 29.8
N2OGWP100 = 273

ghgs = pd.read_excel('non co2 ghgs country.xlsx')
```

```python
def calc_ratio(ghgs, var):
    val_2020 = ghgs[ghgs['Variable']== var]['2020']
    val_2100 = ghgs[ghgs['Variable']== var]['2100']
    ratio = val_2100/val_2020
    return {
        'p50' : ratio.quantile(0.5),
        'p16' : ratio.quantile(0.16),
        'p84' : ratio.quantile(0.84),
    }

def calc_co2eq(ghgs, quantile):
    ch4_2020 = ghgs[ghgs['Variable']== 'Emissions|CH4']['2020'].quantile(0.5) * 10**6 #given in Mt
    n2o_2020 = ghgs[ghgs['Variable']== 'Emissions|N2O']['2020'].quantile(0.5) * 10**3 #given in kt
    ch4_2100 = ch4_2020 * calc_ratio(ghgs, 'Emissions|CH4')[quantile] * CH4GWP100
    n2o_2100 = n2o_2020 * calc_ratio(ghgs, 'Emissions|N2O')[quantile] * N2OGWP100
    return {
        'total_co2eq' : ch4_2100 + n2o_2100,
        'ch4_co2eq' : ch4_2100,
        'n2o_co2eq' : n2o_2100,
        'ch4_2020' : ch4_2020 * CH4GWP100,
        'n2o_2020' : n2o_2020 * N2OGWP100,
    }

def country_compensatory_cdr(ghgs):
    region, ghg, cdr_p50, cdr_p16, cdr_p84, val_2020, model_runs, model_num = [], [], [], [], [], [], [], []

    for country in ghgs['Region'].unique():
        ghg_c = ghgs[ghgs['Region']==country]
        p50 = calc_co2eq(ghg_c, 'p50')
        p16 = calc_co2eq(ghg_c, 'p16')
        p84 = calc_co2eq(ghg_c, 'p84')
        for gas in ['total_co2eq', 'ch4_co2eq', 'n2o_co2eq']:
            region.append(country)
            ghg.append(gas)
            cdr_p50.append(p50[gas]/10**6)
            cdr_p16.append(p16[gas]/10**6)
            cdr_p84.append(p84[gas]/10**6)
            if gas == 'total_co2eq':
                val_2020.append((p50['ch4_2020'] + p50['n2o_2020'])/10**6)
            elif gas == 'ch4_co2eq':
                val_2020.append(p50['ch4_2020']/10**6)
            elif gas == 'n2o_co2eq':
                val_2020.append(p50['n2o_2020']/10**6)
            model_runs.append(len(ghg_c[ghgs['Variable']=='Emissions|CH4']['Scenario']))
            model_num.append(len(ghg_c[ghgs['Variable']=='Emissions|CH4']['Model'].unique()))
    
    results = pd.DataFrame({'region'  : region,
                            'ghg'     : ghg,
                            'p50' : cdr_p50,
                            'p16' : cdr_p16,
                            'p84' : cdr_p84,
                            '2020_value' : val_2020,
                            'model_runs' : model_runs,
                            'model_num' : model_num})
    return results.round(2)
```

```python
country_compensatory_cdr(ghgs)
```
